```python
# Importing the necessary library(s) and gathering the data

import pandas as pd
import matplotlib.pyplot as plt

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

filepath = 'phonepe-pulse_raw-data_q12018-to-q22021-v0-1-5-1720351752.xlsx'

# 0. filepath = phonepe-pulse_raw-data_q12018-to-q22021-v0-1-5-1720351752.xlsx
# 1. sheet_name = State_Txn and Users
# 2. sheet_name = State_TxnSplit
# 3. sheet_name = State_DeviceData
# 4. sheet_name = District_Txn and Users
# 5. sheet_name = District Demographics
```

```python
# TASK 1 - DATA LOADING AND UNDERSTANDING
```

```python
# 1.1 - 1 Loading the data and displaing the top 5 rows

state_txn_and_users = pd.read_excel(filepath, sheet_name = 'State_Txn and Users')
print("Top 5 rows of State_Txn and Users: \n\n",state_txn_and_users.head(5))
```

```python
# 1.1 - 2 Loading the data and displaing the last 10 rows

state_txn_split = pd.read_excel(filepath, sheet_name = 'State_TxnSplit')
print("Last 10 rows of State_TxnSplit: \n\n", state_txn_split.tail(10))
```

```python
# 1.1 - 3 Loading the data and displaing the 10 rows from the middle of the dataset

state_device_data = pd.read_excel(filepath, sheet_name = 'State_DeviceData')

n = len(state_txn_split)
mid = n // 2

mid_10 = state_device_data.iloc[mid - 5 : mid + 5]

print("10 rows from the middle of State_DeviceData: \n\n", mid_10)
```

```python
#  1.1 - 5 Loading the data and displaing the every 10th row of the dataset

district_demographics = pd.read_excel(filepath, sheet_name = 'District Demographics')

print("Every 10th row of District Demographics: \n\n", district_demographics.iloc[::10])
```

```python
# 1.2 - 1 Displaying the summary statistics of 5 datasets

print("SUMMARY STATISTICS \n")
print("State_Txn and Users: \n\n", state_txn_and_users.describe(), "\n\n")
print("State_TxnSplit: \n\n", state_txn_split.describe(), "\n\n")
print("State_DeviceData: \n\n", state_device_data.describe(), "\n\n")
print("District_Txn and Users: \n\n", district_txn_and_users.describe(), '\n\n')
print("District Demographics: \n\n", district_demographics.describe())
```

```python
# 1.2 - 2 Displaying the Data Types of the 5 datasets

print("The data types of each column: State_Txn and Users\n\n", state_txn_and_users.dtypes, "\n\n")
print("The data types of each column: State_TxnSplit \n\n", state_txn_split.dtypes, "\n\n")
print("The data types of each column: State_DeviceData \n\n", state_device_data.dtypes, "\n\n")
print("The data types of each column: District_Txn and Users \n\n", district_txn_and_users.dtypes, "\n\n")
print("The data types of each column: District Demographics \n\n", district_demographics.dtypes)
```

```python
# 1.3 - 1,2,3 Checking for Missing Values

# List of sheet names
sheets = ["State_Txn and Users", "State_TxnSplit", "State_DeviceData", "District_Txn and Users", "District Demographics"]
    
def check_missing(df, name):
    print(f"\nMissing Values in Sheet: {name}")
    missing_count = df.isnull().sum()
    missing_percent = (df.isnull().mean() * 100).round(2)
    missing_summary = pd.DataFrame({
        'Missing Count': missing_count,
        'Missing %': missing_percent})
    if missing_count.sum() == 0:
        print("No missing values found.")
    else:
        max_missing_col = missing_summary['Missing %'].idxmax()
        max_missing_val = missing_summary.loc[max_missing_col, 'Missing %']
        
        print(missing_summary[missing_summary['Missing Count'] > 0])
        print("\n* Column with highest missing %: {max_missing_col} ({max_missing_val}%) *")

# Loop through each sheet
for sheet in sheets:
    df = pd.read_excel(filepath, sheet_name = sheet)
    check_missing(df, sheet)
```

```python
# 1.4 - 1,2 Create a Summary


# Calculating total no. of States & Districts
total_states = state_txn_and_users["State"].nunique()
total_districts = district_txn_and_users["District"].nunique()

# State with Highest Number of Districts
state_district_counts = district_txn_and_users["State"].value_counts()
state_with_max_districts = state_district_counts.idxmax()
max_district_count = state_district_counts.max()

# Displaying Summary
print("SUMMARY")
print(f"Total Number of States: {total_states}")
print(f"Total Number of Districts: {total_districts}")
print(f"State with Highest Number of Districts: {state_with_max_districts} ({max_district_count} districts)")
```

```python
# TASK 2 - EXPLORATORY DATA ANALYSIS (EDA)
```

```python
# 2.1 - 1 Calculating the Total Transactions and Amount (INR) per State over the Years

state_year_summary = (state_txn_and_users.groupby(['State', 'Year'], as_index=False)
    .agg({'Transactions': 'sum','Amount (INR)': 'sum'}).sort_values(['State', 'Year']))

print("Total Transactions and Transaction Amount per State over the Years:\n\n", state_year_summary.head())
```

```python
# 2.1 - 2 Calculating the Top 5 & Bottom 5 States over the Transactions

state_total_txn = (state_txn_and_users.groupby('State', as_index=False)['Transactions'].sum()
    .sort_values('Transactions', ascending=False))

top5_states = state_total_txn.head(5)
bottom5_states = state_total_txn.tail(5)

print("* Top 5 States with Highest Transaction Volumes:\n\n", top5_states.to_string(index=False))
print("\n* Bottom 5 States with Lowest Transaction Volumes:\n\n", bottom5_states.to_string(index=False))
```

```python
# 2.2 - 1 Identifing the most frequent transaction type

# For each state and quarter, find the most frequent transaction type
most_common_txn_type = (state_txn_split.groupby(['State', 'Quarter', 'Transaction Type'], as_index=False)['Transactions'].sum())

# Sort by count descending within each State–Quarter group
most_common_txn_type = (most_common_txn_type.sort_values(['State', 'Quarter', 'Transactions'], ascending=[True, True, False]))

# Pick top transaction type per State–Quarter
top_txn_type_per_state_quarter = (most_common_txn_type.drop_duplicates(subset=['State', 'Quarter'], keep='first').reset_index(drop=True))

# Display results
print("Most Common Transaction Type in Each State and Quarter:\n\n", top_txn_type_per_state_quarter.head())
```

```python
# 2.3 - 1 Determining the device brand with the highest number of registered users in each state

brand_summary = (state_device_data.groupby(['State', 'Brand'], as_index=False)['Registered Users'].sum())

brand_summary = brand_summary.sort_values(['State', 'Registered Users'], ascending=[True, False])

top_brand_per_state = brand_summary.drop_duplicates(subset=['State'], keep='first').reset_index(drop=True)

print("Device Brand with the Highest Number of Registered Users in Each State:\n\n", top_brand_per_state.head())
```

```python
# 2.4 - 1 Identifing the districts with highest population for each state

top_districts = (district_demographics.sort_values(['State', 'Population'], ascending=[False, False])
    .drop_duplicates(subset=['State'],keep='first').reset_index(drop=True))

print("District with the Highest Population in Each State:\n\n", top_districts[['State', 'District', 'Population']].head())
```

```python
# 2.4 - 2 Column Chart

plt.figure(figsize = (16, 8))
plt.bar(top_districts['State'], top_districts['Population'])
plt.title("District with Highest Population in Each State", fontsize=16)
plt.xlabel("State", fontsize = 12)
plt.ylabel("Population", fontsize = 12)
plt.xticks(rotation = 90, fontsize = 9)
plt.tight_layout()
plt.show()
```

```python
# 2.5 - 1, 2 Calculating the average transaction value (ATV) for each state

avg_txn_value = (state_txn_and_users.groupby('State')['ATV (INR)'].mean().sort_values(ascending = False))

top_5_states = avg_txn_value.head(5)
lowest_5_states = avg_txn_value.tail(5)

print("* The average transaction value for each state:\n\n", avg_txn_value, '\n')
print("\n* The top 5 states with the highest ATV:\n\n", top_5_states, '\n')
print("\n* The top 5 states with the lowest ATV:\n\n", lowest_5_states)
```

```python
# 2.6 - 1 Analyzing the app usage trends

app_usage_trends = (state_txn_and_users.groupby(['State', 'Year', 'Quarter'])['App Opens'].sum().reset_index().sort_values(['App Opens'],
                                                                                                                    ascending = False))
print("* Total Number of App Opens Over the Years and Quarters for Each State:\n\n", app_usage_trends)
```

```python
# 2.6 - 2 Ploting the app usage trends

# Select one state for trend visualization
selected_state = "West Bengal"

state_trend = app_usage_trends[app_usage_trends['State'] == selected_state]

# Create a "time" column combining year & quarter for x-axis
state_trend['Time'] = 'Q' + state_trend['Quarter'].astype(str) + ' - ' + state_trend['Year'].astype(str)

# Plot line chart
plt.figure(figsize=(10,5))
plt.plot(state_trend['Time'], state_trend['App Opens'], marker='o', linewidth=2)
plt.title(f'App Usage Trend Over Time – {selected_state}')
plt.xlabel('Year - Quarter')
plt.ylabel('Total App Opens')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 2.7 - 1 Determining the distribution of transaction types

# Identify the most recent quarter
latest_year = state_txn_split['Year'].max()
latest_quarter = state_txn_split[state_txn_split['Year'] == latest_year]['Quarter'].max()

# Filter data for the most recent quarter
recent_data = state_txn_split[(state_txn_split['Year'] == latest_year) & (state_txn_split['Quarter'] == latest_quarter)]

# Groupby State and Transaction Type to calculate total transaction count
txn_distribution = recent_data.groupby(['State', 'Transaction Type'])['Transactions'].sum().reset_index()

pivot_data = txn_distribution.pivot(index = 'State', columns = 'Transaction Type', values = 'Transactions').fillna(0)

# Plotting the stacked bar chart
pivot_data.plot(kind = 'bar', stacked = True, figsize = (16, 10))
plt.title(f'Distribution of Transaction Types per State - Q{latest_quarter}, {latest_year}', fontsize = 14)
plt.xlabel('State', fontsize = 12)
plt.ylabel('Total Transaction Count', fontsize = 12)
plt.xticks(rotation = 90)
plt.legend(title = 'Transaction Type')
plt.tight_layout()
plt.show()
```

```python
# 2.8 - 1,2,3 Find unique mapping between district name and district code

# Extracting only the relevant columns
district_mapping = district_demographics[['District', 'Code']]

unique_district_mapping = district_mapping.drop_duplicates()

# Displaying the unique mappings
print('The unique district mapping: \n', unique_district_mapping.head())

# Exporting to CSV
unique_district_mapping.to_csv("Unique_District_Mapping.csv", index = False)

print("\nUnique district name and code mappings saved to 'Unique_District_Mapping.csv'")
```

```python
# TASK 3 - Data Quality Checks
```

```python
# 3.1 - 1,2,3 Ensuring data consistency across state and district levels

# Aggregate district-level data to state level
district_agg = (district_txn_and_users.groupby(['State', 'Year', 'Quarter'], as_index = False).agg({'Transactions': 'sum',
                                                                                        'Amount (INR)': 'sum','Registered Users': 'sum'}))

# Merging district-level aggregates with state-level data
comparison_df = pd.merge(district_agg, state_txn_and_users[['State', 'Year', 'Quarter', 'Transactions', 'Amount (INR)',
                                'Registered Users']],on = ['State', 'Year', 'Quarter'], how = 'outer', suffixes = ('_district', '_state'))

# Calculating differences
comparison_df['Transactions_Diff'] = comparison_df['Transactions_district'] - comparison_df['Transactions_state']
comparison_df['Amount_Diff'] = comparison_df['Amount (INR)_district'] - comparison_df['Amount (INR)_state']
comparison_df['Users_Diff'] = comparison_df['Registered Users_district'] - comparison_df['Registered Users_state']

# Identifying discrepancies
discrepancies = comparison_df[(comparison_df['Transactions_Diff'] != 0) | (comparison_df['Amount_Diff'] != 0) |
    (comparison_df['Users_Diff'] != 0)]

# Displaying results
if discrepancies.empty:
    print("All state-level data matches district-level aggregates. No discrepancies found.")
else:
    print("Discrepancies found between district-level and state-level data:\n")
    display(discrepancies[['State', 'Year', 'Quarter', 'Transactions_Diff', 'Amount_Diff', 'Users_Diff']])
```

```python
# TASK 4 Data Merging & Advanced Analysis
```

```python
# 4.1 - 1,2 Ratio of users to population by state

# Merging state_txn and users with district demographics
merged_state_districtdemographics = pd.merge(state_txn_and_users, district_demographics, on = 'State', how = 'inner')

# Calculating the ratio of users to population
merged_state_districtdemographics['Users to Population Ratio'] = (merged_state_districtdemographics['Registered Users'] / 
                                                                  merged_state_districtdemographics['Population'])

ratio_by_state = (merged_state_districtdemographics.groupby('State', as_index = False)['Users to Population Ratio'].mean())

print("\nRatio of Registered Users to Population by State:\n\n", ratio_by_state, '\n')


plt.figure(figsize = (14,7))
plt.bar(ratio_by_state['State'], ratio_by_state['Users to Population Ratio'])
plt.title('Ratio of Registered Users to Population by State', fontsize = 14)
plt.xlabel('State')
plt.ylabel('User to Population Ratio')
plt.xticks(rotation = 90)
plt.tight_layout()
plt.show()
```

```python
# 4.2 - 1,2,3 Correlate population density with transaction volume

# Merging District_Txn and Users on District Demographics
merged_district = pd.merge(district_txn_and_users, district_demographics, on = 'District', how = 'inner')

# Calculating correlation between population density and transaction volume
correlation = merged_district['Population'].corr(merged_district['Transactions'])

print(f"\nCorrelation between Population Density and Transaction Volume: {correlation:.4f}, \n\n")

# Scatter plot to visualize correlation
plt.figure(figsize = (10,6))
plt.scatter(merged_district['Population'], merged_district['Transactions'], alpha = 0.6)
plt.title('Correlation: Population Density vs Transaction Volume', fontsize = 14)
plt.xlabel('Population Density')
plt.ylabel('Transaction Volume')
plt.grid(True, linestyle = '--', alpha = 0.5)
plt.tight_layout()
plt.show()
```

```python
# 4.3 - 1,2 Average transaction amount per user

# Merging state_txn_and_users datasets on district_demographics
merged_state = pd.merge(state_txn_and_users, district_demographics, on='State', how='inner')

# Calculating Average Transaction Amount per User
merged_state['Avg_Transaction_per_User'] = (merged_state['Amount (INR)'] / merged_state['Registered Users'])

# Group by State to get state-level averages (in case of multiple entries)
avg_txn_per_user = (merged_state.groupby('State', as_index=False)['Avg_Transaction_per_User'].mean())

print("\n* Average Transaction Amount per User by State:\n\n", avg_txn_per_user)

top5_highest = avg_txn_per_user.sort_values(by='Avg_Transaction_per_User', ascending=False).head(5)

top5_lowest = avg_txn_per_user.sort_values(by='Avg_Transaction_per_User', ascending=True).head(5)

print("\n* Top 5 States with Highest Average Transaction Amount per User:\n\n", top5_highest, "\n\n")
print("\n* Top 5 States with Lowest Average Transaction Amount per User:\n\n", top5_lowest)
```

```python
# 4.4 - 1,2,3 Device brand usage ratio

# Merging State_DeviceData with State_Txn_and_Users
merged_SDD_STU = pd.merge(state_device_data, state_txn_and_users, on = 'State', how = 'inner', suffixes = ('_SDD', '_STU'))

# Calculating device brand usage ratio
merged_SDD_STU['Device_Brand_Usage_Ratio'] = (merged_SDD_STU['Registered Users_SDD'] / merged_SDD_STU['Registered Users_STU'])

# Relevant columns for display
device_usage_ratio = merged_SDD_STU[['State', 'Brand', 'Device_Brand_Usage_Ratio']]

print("\nDevice Brand Usage Ratio by State:\n\n", device_usage_ratio.head(10), '\n\n')

Selected_State = 'West Bengal' #input()

# Visualization - Bar chart for each state
plt.figure(figsize = (14,7)) 
for state in device_usage_ratio['State'].unique():
    subset = device_usage_ratio[device_usage_ratio['State'] == Selected_State]
plt.bar(subset['Brand'], subset['Device_Brand_Usage_Ratio'], label = Selected_State)
plt.title("Device Brand Usage Ratio by State", fontsize = 14)
plt.xlabel("Device Brand")
plt.ylabel("Usage Ratio")
plt.legend(title = "Selected_State", bbox_to_anchor = (1.05, 1), loc = 'upper left')
plt.tight_layout()
plt.show()
```

```python
# 5.2 - 1 Create a pie chart showing the distribution of transaction types for a specific quarter

# Select state and quarter for analysis
selected_state = "West Bengal"
selected_quarter = 2
selected_year = 2021

# Filtering dataset
state_quarter_data = state_txn_split[(state_txn_split['State'] == selected_state) & (state_txn_split['Quarter'] == selected_quarter) &
    (state_txn_split['Year'] == selected_year)]

# Group by transaction type
txn_type_distribution = (state_quarter_data.groupby('Transaction Type')['Transactions'].sum().reset_index())

# Pie chart
plt.figure(figsize=(6,6))
plt.pie(txn_type_distribution['Transactions'], labels = txn_type_distribution['Transaction Type'], autopct = '%1.2f%%', startangle = 90,
    wedgeprops = {'edgecolor': 'black'})
plt.title(f"Transaction Type Distribution - {selected_state}, Q{selected_quarter} {selected_year}", fontsize = 14)
plt.show()
```

```python
# 5.3 - 1 Create a bar plot showing the population density of districts in a selected state.

selected_state = 'West Bengal'

# Filtering data for selected state
state_data = district_demographics[district_demographics["State"] == selected_state]

# Calculating Population Density
state_data["Population Density"] = state_data["Population"] / state_data["Area (sq km)"]

plt.figure(figsize = (12,6))
plt.bar(state_data["District"], state_data["Population Density"], color = 'skyblue', edgecolor = 'black')
plt.xticks(rotation = 75, ha = 'right')
plt.title(f"Population Density by District in {selected_state}", fontsize = 14)
plt.xlabel("District")
plt.ylabel("Population Density (people per sq. km)")
plt.tight_layout()
plt.show()
```

```python
# 6.1 - 1 Identify any trends or patterns in the transaction data

# Grouping by Year and Quarter to find trends
txn_trends = state_txn_and_users.groupby(["Year", "Quarter"], as_index=False).agg({"Transactions": "sum","Amount (INR)": "sum"})

# Plotting total transactions and amount trends
plt.figure(figsize = (10,5))
plt.plot(txn_trends["Year"].astype(str) + "-Q" + txn_trends["Quarter"].astype(str),txn_trends["Transactions"], marker = 'o',
         label = "Total Transactions")
plt.plot(txn_trends["Year"].astype(str) + "-Q" + txn_trends["Quarter"].astype(str),txn_trends["Amount (INR)"], marker = 's', 
         label = "Total Transaction Amount (INR)")
plt.title("Trend in Transactions Over Time (Year-Quarter)")
plt.xlabel("Year-Quarter")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print("""
Findings (6.1 - 1):
-------------------
- Transactions and total amounts show consistent growth over time.
- Significant surge visible post-2020, likely due to increasing digital adoption.
- Seasonal peaks may align with festive or high-purchase quarters.
""")
```

```python
# 6.2 - 1 Correlate Demographic Data with Transaction Data

# Calculating population density
district_demographics["Population Density"] = district_demographics["Population"] / district_demographics["Area (sq km)"]

# Merging datasets on District and State
merged_df = pd.merge(district_txn_and_users, district_demographics, on=["State", "District"], how="inner")

# Step 5: Calculate correlation between population density and transactions
correlation = merged_df[["Population Density", "Transactions", "Amount (INR)"]].corr()

print("Correlation between Demographics and Transactions:\n--------------------------------------------------\n", correlation)

print("""
Findings (6.2 - 1):
-------------------
- A positive correlation between population density and transaction volume suggests that densely populated areas use more digital payments
- However, correlation may not be perfect, as factors like income level and smartphone penetration also play a role.
""")
```
